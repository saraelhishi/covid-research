```python
tweets['text_lemma1']=tweets['text_lemma1'].apply(lambda x:' '.join([word for word in str(x).split() if word!= '-PRON-']))
```

```python
# Sentiment Analysis

#keyword = input("Pease enter keywords or hashtags you search for:")
#noOfTweets = int(input("Enter number of tweets:"))
# scrape
#tweets = tweepy.Cursor(api.search, q=keyword).items(noOfTweets)
noOfTweets=len(tweets)
positive=0
negative=0
neutral=0
polarity=0
tweet_list=[]
neutral_list=[]
positive_list=[]
negative_List=[]

for tweet in tweets.text_clean:
    #print(tweet.text)
    tweet_list.append(tweet)
    analysis=TextBlob(tweet)
    score=SentimentIntensityAnalyzer().polarity_scores(tweet)
    neg=score['neg']
    neu=score['neu']
    pos=score['pos']
    comp=score['compound']
    
    polarity += analysis.sentiment.polarity
    if neg > pos:
        negative_List.append(tweet)
        negative += 1
    elif pos > neg:
        positive_list.append(tweet)
        positive += 1
    elif pos == neg:
        neutral_list.append(tweet)
        neutral += 1
# cacluate percentage
def percentage(part, whole):
    return 100 * float(part)/float(whole)
positive = format(percentage(positive, noOfTweets), '.1f')
negative = format(percentage(negative, noOfTweets), '.1f')
neutral = format(percentage(neutral, noOfTweets), '.1f')
polarity=percentage(polarity, noOfTweets)
```

```python
#Number of Tweets (Total, Positive, Negative, Neutral)
tweet_list = pd.DataFrame(tweet_list)
neutral_list = pd.DataFrame(neutral_list)
negative_List = pd.DataFrame(negative_List)
positive_list = pd.DataFrame(positive_list)
print("total number: ",len(tweet_list))
print("positive number: ",len(positive_list), "(",positive,"%)")
print("negative number: ", len(negative_List), "(",negative,"%)")
print("neutral number: ",len(neutral_list), "(",neutral,"%)")
```

```python
# Creating Pie Chart
labels=['Positive ['+str(positive)+'%]', 'Neutral ['+str(neutral)+'%]', 'Negative ['+str(negative)+'%]']
sizes=[positive, neutral, negative]

patches, texts = plt.pie(sizes, startangle=90)
plt.style.use('default')
plt.legend(labels)
#plt.title("Sentiment Analysis Result for keyword= "+keyword)
plt.axis('equal')
plt.show()
```

```python
#### Sentiment Analysis of Cleaned dataset
# correct the df shape
#df = pd.DataFrame(tw_df)
#df.drop(columns=['cleaned_text','tweet'], inplace=True)
#Calculating Negative, Positive, Neutral and Compound values
tweets[['polarity','subjectivity']]= tweets['text_clean'].apply(lambda Text: pd.Series(TextBlob(Text).sentiment))

for index, row in tweets['text_clean'].iteritems():
    score = SentimentIntensityAnalyzer().polarity_scores(row)
    neg=score['neg']
    neu=score['neu']
    pos=score['pos']
    comp=score['compound']
    
    if neg > pos:
        tweets.loc[index,'sentiment']='negative'
    elif pos > neg:
        tweets.loc[index,'sentiment']='positive'
    else:
        tweets.loc[index,'sentiment']='neuteral'
    tweets.loc[index,'neg']=neg
    tweets.loc[index,'pos']=pos
    tweets.loc[index,'neu']=neu

# print 
tweets.head(10)
```

```python
#Function for count_values_in single columns
def count_values_in_column(data, feature):
    total=data.loc[:,feature].value_counts(dropna=False)
    percentage= round(data.loc[:,feature].value_counts(dropna=False,normalize=True)*100,2)
    return pd.concat([total, percentage], axis=1, keys=['Total','Percentage'])

# count values for sentiment
count_values_in_column(tweets,'sentiment')
```

```python
# create data for Pie Chart
pichart=count_values_in_column(tweets,'sentiment')
names=pichart.index
size=pichart['Percentage']

# Create a circle for the center of the plot
circle=plt.Circle((0,0), 0.7, color="white")
plt.pie(size, labels=names)
p=plt.gcf()
p.gca().add_artist(circle)
plt.show()
```

```python
#Function to Create Wordcloud
def create_wordcloud(text):
    #mask=np.array(Image.open("images//tweet-mask.png"))
    stopwords=set(STOPWORDS)
    #wc= WordCloud(background_color="white",mask=mask, max_words=3000, stopwords=stopwords, repeat=True)
    wc= WordCloud(background_color="white",max_words=3000, stopwords=stopwords, repeat=True)
    wc.generate(str(text))
    #wc.to_file("wc.png")
    #print("Word Cloud Saved!")
   # path="../wc.png"
   # display(Image.open(path))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```
